```python
import seaborn as sns
```

```python
df=sns.load_dataset("titanic") #titanic bcz yahan par bohut sara missing values hai
```

```python
df.head()
```

```python
#agar hame missing values check karna hai ki woh kahan par hai
#jahan par missing values hoga woh true hoga aaur jaha par nahin hoga woh false hoga
df.isnull()
```

```python
#agar hame pata karna hai ki total kitne missing values hai har column main to hum usse by sum pata kar sakte hein
df.isnull().sum()
```

```python
sns.heatmap(df.isnull()) #agar hame upar wale cheez ko graphically dekhna hai
```

```python
#handling missing values can be done by deleting the rows
df.dropna()  #yeah kya karta hai jahan jahan nan value hai uss row ko delete kardeta hai
#agar koi parameter na dein toh jis jis row mein nan values hai un sabko delete kar deta hai
```

```python
df.dropna().shape
```

```python
#Handling missing values by deleting columns
df.dropna(axis=1)
```

```python
#imputation techniques
#1-Mean value imputation
#this technique works well when your data is normally distributed
```

```python
sns.displot(df["age"])
```

```python
df.age.isnull().sum() #these are the missing values
```

```python
df["Age_mean"]=df["age"].fillna(df['age'].mean())
```

```python
df[['Age_mean','age']]
```

```python
#2=meadian value imputation
df['Age_median']=df['age'].fillna(df['age'].median())
```
